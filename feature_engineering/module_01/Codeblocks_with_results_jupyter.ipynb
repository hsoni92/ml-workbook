```python
# STEP 1: Create a simple DataFrame to simulate student marks in two subjects
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# Create a basic DataFrame
df_easy = pd.DataFrame({
'Maths': [35, 67, 80, 95, 42],
'Science': [45, 76, 88, 90, 39],
'Passed': [0, 1, 1, 1, 0] # Target: 1 = Pass, 0 = Fail
})

# STEP 2: Add a new feature – Average Marks
df_easy['Average'] = df_easy[['Maths', 'Science']].mean(axis=1)

# View the new DataFrame
print("Basic Marks Dataset with Engineered 'Average' Feature:")
print(df_easy)

# STEP 3: Visualize Average Marks vs Passed
plt.figure(figsize=(6, 4))
sns.barplot(x='Passed', y='Average', data=df_easy, palette='Set2')
plt.title("Average Marks by Pass/Fail")
plt.xlabel("Passed (0 = No, 1 = Yes)")
plt.ylabel("Average Marks")
plt.tight_layout()
plt.show()

# Summary:
print("✅ We added a new feature 'Average' and found that passing students have clearly higher average scores.")
```

Level 2 IRIS Dataset

Block 1: Load Required Libraries

```python
# Import necessary libraries
from sklearn.datasets import load_iris  # To load the inbuilt Iris dataset
import pandas as pd                     # For data manipulation and DataFrame
import numpy as np                      # For numerical operations
import matplotlib.pyplot as plt         # For plotting
import seaborn as sns                   # For advanced visualizations
from sklearn.ensemble import RandomForestClassifier  # For feature importance
#This block loads all the Python libraries needed for data loading, exploration, visualization, and modeling.
```

Block 2: Load and Explore the Iris Dataset

```python
# Load the inbuilt Iris dataset
iris = load_iris()

# Create a DataFrame from the dataset features
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target (species) as a new column with readable names
df_iris['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Display the first 5 rows of the dataset
print("Iris Dataset (First 5 Rows):")
print(df_iris.head())
#This block loads the Iris dataset, converts it into a readable DataFrame, and shows the first few rows to understand the structure.
```

Block 3: Melt the Data for Multi-Feature Plotting

```python
# Melt the dataset to long format for easier plotting with seaborn
df_melted = df_iris.melt(id_vars='species', var_name='Feature', value_name='Value')

# Display the melted DataFrame
print("Melted Data (First 5 Rows):")
print(df_melted.head())
#This transforms the wide-format DataFrame into a long-format, which is more suitable for seaborn plots (like violin or boxplots).
```

Block 4: Violin Plot for Feature Distributions by Species

```python
# Plot violin plots for each feature by species
plt.figure(figsize=(12, 6))
sns.violinplot(x="Feature", y="Value", hue="species", data=df_melted, palette='muted')
plt.title("Feature Distributions by Iris Species")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#This shows the distribution of each feature for different species using a violin plot. Helps identify which features separate species well.
```

Block 5: Train Random Forest Classifier

```python
# Prepare the features (X) and target (y) for the model
X = df_iris[iris.feature_names]
y = df_iris['species']

# Train a Random Forest Classifier to find feature importance
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)
#This block trains a Random Forest model on the Iris dataset to learn which features are most important for classifying the species.
```

Block 6: Plot Feature Importances

```python
# Extract and sort feature importance values
importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(8, 5))
sns.barplot(x='Importance', y='Feature', data=importances, palette='rocket')
plt.title("Feature Importance (Iris Dataset)")
plt.tight_layout()
plt.show()
#This visualizes the importance of each feature as calculated by the Random Forest model. More importance = more predictive power.
```

Block 7: Final Summary

```python
# Print the final summary based on the visualizations and model
print("From the violin plots and Random Forest results:")
print("- Petal length and petal width are the most important features for predicting Iris species.")
#This summarizes the learning: Which features matter the most in identifying the flower species.
```

Level 3 Wine Quality Dataset

Block 1: Load Required Libraries

```python
# Import required Python libraries for data handling and visualization
import pandas as pd                # For loading and handling dataset
import seaborn as sns              # For plotting heatmaps and boxplots
import matplotlib.pyplot as plt    # For customizing visualizations
```

Block 2: Load the Wine Quality Dataset

```python
# Load the Red Wine Quality dataset from local system
wine = pd.read_csv("C:/Users/mahip/Downloads/winequality-red.csv")  # Ensure the file path is correct

# Check the shape (rows and columns) and preview first 5 rows
print("Wine Dataset Shape:", wine.shape)
print(wine.head())
#This block loads the dataset and shows its structure. Helps identify the number of samples and features.
```

Block 3: Check for Missing Values

```python
# Check if any feature/column has missing values
print("Null Values in Each Column:\n", wine.isnull().sum())
# Important for feature engineering: If null values exist, we must impute or drop them before modeling.
```

Block 4: Correlation Heatmap (Feature Selection)

```python
# Visualize correlation between features using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(wine.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap - Wine Features")
plt.show()
#This heatmap shows how strongly features are related to each other and to the target quality. Helps decide which features are useful or redundant.
```

Block 5: Boxplots – Visualize Key Feature Relationships with Target

```python
# Visual analysis of how key features vary across wine quality
key_features = ['alcohol', 'volatile acidity', 'citric acid']  # Based on domain knowledge or correlation

# Loop through each feature and plot it against wine quality
for feature in key_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='quality', y=feature, data=wine, palette='coolwarm')
    plt.title(f"{feature} vs Wine Quality")
    plt.tight_layout()
    plt.show()
#Boxplots help identify how each key feature behaves across different quality ratings. Useful for understanding predictive strength and outliers.
```

Block 6: Prepare Data for Feature Importance using Random Forest

```python
# Import modeling tools
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#These tools help with preprocessing (scaling), splitting the data, and training a Random Forest classifier to measure feature importance.
```

Block 7: Data Preparation (Splitting and Scaling)

```python
# Separate input features (X) and target label (y)
X = wine.drop('quality', axis=1)  # Drop target variable
y = wine['quality']               # Target variable

# Standardize the features (important for tree-based models if interpreting results)
X_scaled = StandardScaler().fit_transform(X)

# Split dataset into training and testing sets (80/20)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
#This block prepares your data for machine learning – clean separation and normalization ensures fair training and interpretation.
```

Block 8: Train the Random Forest Model

```python
# Train a Random Forest Classifier to find most predictive features
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
#Random Forest helps identify which features most strongly impact wine quality through its internal feature importance metric.
```

Block 9: Plot Feature Importance

```python
# Create a DataFrame of feature importances from the trained model
importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Plot the importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importances, palette='mako')
plt.title("Feature Importance - Wine Dataset")
plt.tight_layout()
plt.show()
#This plot ranks features based on how much they contribute to predicting wine quality. Higher importance = more predictive power.
```

Block 10: Final Summary

```python
# Print takeaway from the visual and model analysis
print("Insights from Feature Engineering:")
print("- Alcohol and sulphates are highly predictive of wine quality.")
print("- Volatile acidity has a strong negative correlation with quality.")
print("- This kind of feature analysis helps select best variables for predictive modeling.")
#This final message summarizes your findings and reinforces why feature engineering is critical before building any ML model.
```
