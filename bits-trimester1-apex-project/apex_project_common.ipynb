# **DATA CLEANING**

```python
#No missing data, No Negative Values, No Duplicate rows (Already Checked in Data Audit)
# DROP IRRELEVANT COLUMNS
datan = data.drop(['ID', 'Locality','Price_per_SqFt','Year_Built','Total_Floors','Facing','Owner_Type'], axis=1)
datan.head()
```

```python
# OUTLIER DETECTION
numeric_col = datan.select_dtypes(include=['int64', 'float64']).columns
outlier_indices = {}

for col in numeric_col:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    outlier_indices[col] = outliers.index.tolist()

    print(f"{col}: {len(outliers)} outliers")
```

##No Outliers Found

# **FEATURE ENGINEERING**

```python
#EXTRACTING NEW FEATURE FROM THE EXISTING FEATURE "AMENITIES"

split_amenities = datan['Amenities'].str.split(',')
# Step 2: Flatten the list of lists into a single list and strip spaces
all_amenities = [item.strip() for sublist in split_amenities for item in sublist]
# Step 3: Get unique items
unique_amenities = set(all_amenities)
unique_amenities
```

```python
datan["Amenities"] = datan["Amenities"].str.replace(r"\s+", "", regex=True).str.lower()
data_encoded = datan["Amenities"].str.get_dummies(sep=",")
datam = pd.concat([datan, data_encoded], axis=1)
datam.drop(columns='Amenities',inplace=True)
datam.shape
```

```python
datam.head()
```

```python
#Using Labelencoder to tranform categorical variables to numeric variables

from sklearn.preprocessing import LabelEncoder
X=datam.drop(columns=['Price_in_Lakhs'],axis=1)
y=datam['Price_in_Lakhs']
cat_f = X.select_dtypes(include='object').columns
X_encoded = X.copy()
for col in cat_f:
    X_encoded[col] = LabelEncoder().fit_transform(X_encoded[col])

X_encoded.head()
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression

selector = SelectKBest(score_func=mutual_info_regression, k=5)
selector.fit(X_encoded, y)

mi_scores = pd.DataFrame({
    'Feature': X_encoded.columns,
    'MI Score': selector.scores_
}).sort_values(by='MI Score', ascending=False)

print(mi_scores.head(23))
```

```python
from sklearn.feature_selection import f_regression

# --- 3️⃣ Apply f_regression ---
f_selector = SelectKBest(score_func=f_regression, k='all')
f_selector.fit(X_encoded, y)

# --- 4️⃣ Collect scores ---
f_scores = pd.DataFrame({
    'Feature': X_encoded.columns,
    'F_Score': f_selector.scores_,
    'p_value': f_selector.pvalues_
}).sort_values(by='F_Score', ascending=False)

print("🔹 Top 10 Features by F-Score (Linear Relation with Price):")
display(f_scores.head(10))
```

```python
# Train model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_encoded, y)

# Get feature importance
importances = pd.Series(rf.feature_importances_, index=X_encoded.columns)
top_features = importances.sort_values(ascending=False).head(15)

print(top_features)
```
