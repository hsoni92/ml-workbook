## Problem Statement & Business Goal

A Portuguese bank used phone calls to offer a term deposit. Each row is one contact; the label `y` is whether the customer subscribed (`yes` / `no`). Calling everyone is expensive, so the bank wants to focus on people more likely to say yes.

For this project I will build a binary classifier and compare a **neural network (MLP in Keras)** with **logistic regression** as a baseline. Because `yes` is the minority class, I will pay more attention to recall, F1, and PR-AUC than to accuracy alone.

## Dataset Details

- **Source:** UCI Bank Marketing — https://archive.ics.uci.edu/dataset/222/bank+marketing  
- **File used:** `data/bank-full.csv` (semicolon-separated)  
- **Size:** 45,211 rows, 17 columns (one row per phone contact)  
- **Target:** `y` — `yes` if the customer took the term deposit, else `no`  
- **Features:** mix of numeric (`age`, `balance`, `duration`, …) and categorical (`job`, `month`, `poutcome`, …)

`duration` is call length in seconds. It is very predictive but only known after the call ends, so I will try models with and without it and note the difference.

## Proposed Workflow
